```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from collections import Counter
```

```python
def nth_permutation(n, idx):
    a = n
    b = n
    res = []

    for _ in range(2 * n):
        if a == 0:
            res.append('0')
            b -= 1
            continue
        if b == 0:
            res.append('1')
            a -= 1
            continue

        cnt = math.comb(a + b - 1, a - 1)

        if idx < cnt:
            res.append('1')
            a -= 1
        else:
            res.append('0')
            idx -= cnt
            b -= 1

    return ''.join(res)
```

```python
def get_cnt (mask):
    A = []
    B = []
    for i, x in enumerate(mask): 
        if x == 1: 
            A.append(i)
        else: 
            B.append(i)
    cnt = {}
    for a in A: 
        for b in B: 
            diff = a-b
            if diff in cnt:
                cnt[diff] += 1
            else:
                cnt[diff] = 1
    return cnt
```

```python
def read_file (filename):
    with open (filename) as f: 
        lines = f.readlines()
        
    df = pd.DataFrame([int(x) for x in lines[6].split()[:-11]])
    ax = df.plot(linewidth=0.05)
    plt.show()
    df.hist()
    return df
```

```python
def score(counts):
    return sum(d * d * c for d, c in counts.items()) * max(counts.keys())
```
